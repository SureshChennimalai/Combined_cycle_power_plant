```python
# define X and Y
X = dataset[col_names[:-1]]
y = dataset[col_names[-1]]
```

```python
dataset.AT.quantile([0.25,0.75])
```

```python
# Remove outliers in RH
IQR = abs(np.subtract(dataset.RH.quantile(0.25),dataset.RH.quantile(0.75)))
lower_whisker = dataset.RH.quantile(0.25) - 1.5*IQR  
upper_whisker = dataset.RH.quantile(0.75) + 1.5*IQR  
#dataset = dataset[(dataset.RH > lower_whisker) & (dataset.RH < upper_whisker)]
```

```python
# Remove outliers in AP
IQR = abs(np.subtract(dataset.AP.quantile(0.25),dataset.AP.quantile(0.75)))
lower_whisker = dataset.AP.quantile(0.25) - 1.5*IQR  
upper_whisker = dataset.AP.quantile(0.75) + 1.5*IQR  
#dataset = dataset[(dataset.AP > lower_whisker) & (dataset.AP < upper_whisker)]
```

```python
# shuffle data and split

(X_train, X_test, y_train, y_test) = train_test_split(X,y, test_size = .3, random_state = 52)
```

```python
# Apply RF
#X, y = make_regression(n_features=4, n_informative=2, random_state=0, shuffle=False)
regr = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
regr.fit(X_train, y_train)

print(regr.feature_importances_)
```

```python
# Predict accuracy
prediction = regr.predict(X_test)
r2_score(y_test, prediction) 
```

```python
import shap

# load JS visualization code to notebook
shap.initjs()
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_test)

# visualize the first prediction's explanation
shap.force_plot(explainer.expected_value, shap_values[0,:], X_test.iloc[0,:])
```
